#### PLAN

1. Makes sense to start with the elite team plus the cheapest reserves available. 
2. We won't worry about any chips in this problem so the bench can just be the cheapest available players in each position at gw2. This is just the opening gambit, we may want to bench more than one player from a particular position - a regular occurence in the game. 
2. We've ordered each position in terms of points scored
3. We'll need to iterate through all the next set of players who just missed the cut from all positions and find the next most valueable player. If he is cheaper than the lowest points scorer in the squad then he is selected, if not then we look again for the next cheapest player across all positions and repeat.

step 1: replace the lowest scorer in each position with the highest scoring player at the cheapest price from each position - this creates the bench\
step 2: look across all positions in players who just missed the cut and find the next highest global scorer. \
step 3: If cheaper than lowest scorer in the team - replace the lowest scorer in the team, in that position with the new player\
step 4: If not cheaper - go back to step 2

### BENCHED TEAM

```python
def benched_players(df):
    df = df
    
    #Â Select the players to be benched using the rules set out in step 1 in the plan
    gk_bench = gk.iloc[1:].sort_values(by="cost")[gk.cost==gk.cost.min()].iloc[-1].id
    def_bench = dfn.iloc[1:].sort_values(by="cost")[dfn.cost==dfn.cost.min()].iloc[-1].id
    mid_bench = mid.iloc[1:].sort_values(by="cost")[mid.cost==mid.cost.min()].iloc[-1].id
    fwd_bench = fwd.iloc[1:].sort_values(by="cost")[fwd.cost==fwd.cost.min()].iloc[-1].id
    
    # now set the bench location in the frame for each position
    gk_loc = 1,0
    def_loc = 4,1
    mid_loc = 4,2
    fwd_loc = 2,3
    
    # Replace the bench positions with our new player names
    df.iloc[gk_loc] = gk_bench
    df.iloc[def_loc] = def_bench
    df.iloc[mid_loc] = mid_bench
    df.iloc[fwd_loc] = fwd_bench
    
    return df
```

```python
# Call the function into a variable
benched_frame = benched_players(team_frame)
```

### Budget Team

**Detailed Selection process**
1. First we'll reorganise the base team so we sort by position (descending( and then by points (ascending)
2. Look through next highest socring player (not currently selected) in each position. 
2. Select the player with the lowest net loss of points from the player he could replace.
3. If he is cheaper than the lowest scoring player in the team, in that position: transfer the player in and move to the next highest scoring player of the 4 and repeat.
4. If he is more expensive: move to the next highest scoring player of the 4.
5. Repeat steps 3 and 4 until all four players eith transfered in or discarded
6. Recalculate the team cost
7. if cost < 100: stop process and return the finalised team
8. if cost > 100: iterate i+=1
9. Repeat process from step 1 until cost < 100

1. Reorder the team as per above rules

```python
# Reindex and sort values by element type asceding and then by cum points descending
base_gk = base_team[base_team.element_type==1].sort_values(by="cum_points", ascending=False)
base_dfn = base_team[base_team.element_type==2].sort_values(by="cum_points", ascending=False)
base_mid = base_team[base_team.element_type==3].sort_values(by="cum_points", ascending=False)
base_fwd = base_team[base_team.element_type==4].sort_values(by="cum_points", ascending=False)

base_team_sorted = pd.DataFrame(columns=base_gk.columns)
base_team_sorted = base_team_sorted.append(base_gk).append(base_dfn).append(base_mid).append(base_fwd)
base_team_sorted.reset_index(inplace=True)
base_team_sorted.drop("index", axis=1, inplace=True)
```

```python
base_team_sorted
```

2. Build a function that returns the best budget team.

```python
budget_team = base_team_sorted
for i in range(gk.shape[0]-3):
    
    g = i+2
    d = i+5
    m = i+5
    f = i+4

    #review_list = []

    # Set up an empty dataframe to store the next players in each position
    players = pd.DataFrame(columns = gk.columns)

    # Store each of the next players in an isolated frame
    next_gk = gk.iloc[g].to_frame().T
    next_def = dfn.iloc[d].to_frame().T
    next_mid = mid.iloc[m].to_frame().T
    next_fwd = fwd.iloc[f].to_frame().T

    # Add the players into the empty dataframe
    players = players.append(next_gk).append(next_def).append(next_mid).append(next_fwd)

    # Check which one would result in the smallest net loss in points to the team, if transfered in

    # Set a list of net losses
    net_loss = []

    # calculate the potential loss each transfer would incur
    gk_loss = budget_team.cum_points.iloc[1] - players.cum_points.iloc[0]
    def_loss = budget_team.cum_points.iloc[6] - players.cum_points.iloc[1]
    mid_loss = budget_team.cum_points.iloc[11] - players.cum_points.iloc[2]
    fwd_loss = budget_team.cum_points.iloc[14] - players.cum_points.iloc[3]

    # Add each loss to the net_loss list which is our unordered list of net losses
    net_loss.extend([gk_loss, def_loss, mid_loss, fwd_loss])

    # Create an ordered list of net losses high to low
    ordered_loss = net_loss.copy()
    ordered_loss.sort()

    #Â Now create an ordered list of indexes - i.e the index in net_loss that we can find values in the ordered_loss
    # Note the values have 1 added to equate to the element types 1=gk, 2=def, 3=mid, 4=fwd
    ordered_idx = [net_loss.index(ordered_loss[x])+1 for x in range(len(ordered_loss))]
    
    # Now we can iterate through the ordered_idx list and apply the value at that index as the index to find the next
    # most productive player in the current set. If they don't pass the cost test then we move onto the next
    
    for i, j in enumerate(ordered_idx):
        if j == 1 and players.iloc[0].cost < budget_team.iloc[1].cost: #index 1 in base team == worst keeper
            budget_team = budget_team.drop(index=1, axis=0)
            budget_team.loc[1] = players.iloc[0]
            budget_team.sort_index(inplace=True)
            
        elif j == 2 and players.iloc[1].cost < budget_team.iloc[6].cost: # index 6 in base team == worst defender
            budget_team= budget_team.drop(index=6, axis=0)
            budget_team.loc[6] = players.iloc[1]
            budget_team.sort_index(inplace=True)
        
        elif j == 3 and players.iloc[2].cost < budget_team.iloc[11].cost: # index 11 in base team == worst midfield
            budget_team = budget_team.drop(index=11, axis=0)
            budget_team.loc[11] = players.iloc[2]
            budget_team.sort_index(inplace=True)
            
        elif j == 4 and players.iloc[3].cost < budget_team.iloc[14].cost: # index 11 in base team == worst midfield
            budget_team = budget_team.drop(index=14, axis=0)
            budget_team.loc[14] = players.iloc[3]
            budget_team.sort_index(inplace=True)
            
            
    # If we have achieved a squad value equal or less than the Â£100M budget then we can break the loop    
    if budget_team.cost.sum()/10 < 100:
        break
    else:
        continue
    


    
    
    
    
```

#### Plan
If a player has scored less than the 75%ile score for points and has lower than the 25%ile score for value they enter a category of player we'll call low_value

```python
# In the base team we have the following 'low' score high cost players >>> low_value
low_value = base_team_sorted[(base_team_sorted.cum_points<163) & (base_team_sorted.value<1.9)]
low_value = low_value.sort_values(by="value", ascending=False)
low_value
```

#### Narrative
We have two player currently in the low value category. Rashford and Vardy. Now Rashford is also the lowest scoring player in his position so he is already in the line of fire for transfer. Vardy, however is not the lowest points scorer in his position, but because of his being in the low value category he will automatically be pushed into the next to be transferred position. 

If there was another forward who had a lower value than Vardy, and in the low_value set, then that player would be next in line. 

#### Summary
With this new rule Bamford and Vardy would swap places and we would be looking to transfer Vardy out for a cheaper player. 

```python
budget_team = base_team_sorted
for i in range(gk.shape[0]-3):
    
    # Set the index at which we're looking for a player to be transfered in from the full set of players
    # ie the next best scoring player not currently in the squad
    g = i+2
    d = i+5
    m = i+5
    f = i+4

    #review_list = []
    
    # Now we need to decide at what index, from the base squad, the player to be transfered is located at. Typically 
    # the indexes are set at the lowest scoring player location, however, with our new value limits this could change!
    
    # Set base indexes
    g_idx, d_idx, m_idx, f_idx = 1, 6, 11, 14
    
    # Now check if these need to be changed
    # First group the base team by position
    positions = base_team_sorted.groupby("element_type")
    
    for j in range(1, 5):
        pos = positions.get_group(j)
        new_transfer = pos[pos.web_name.isin(low_value.web_name)]
        if j == 1 and new_transfer.shape[0]>0:
            g_idx = new_transfer.index.values[0]
        elif j == 2 and new_transfer.shape[0]>0:
            d_idx = new_transfer.index.values[0]
        elif j == 3 and new_transfer.shape[0]>0:
            m_idx = new_transfer.index.values[0]
        elif j == 4 and new_transfer.shape[0]>0:
            f_idx = new_transfer.index.values[0]

    # Set up an empty dataframe to store the next players in each position
    players = pd.DataFrame(columns = gk.columns)

    # Store each of the next players in an isolated frame
    next_gk = gk.iloc[g].to_frame().T
    next_def = dfn.iloc[d].to_frame().T
    next_mid = mid.iloc[m].to_frame().T
    next_fwd = fwd.iloc[f].to_frame().T

    # Add the players into the empty dataframe
    players = players.append(next_gk).append(next_def).append(next_mid).append(next_fwd)

    # Check which one would result in the smallest net loss in points to the team, if transfered in

    # Set a list of net losses
    net_loss = []

    # calculate the potential loss each transfer would incur
    gk_loss = budget_team.cum_points.iloc[g_idx] - players.cum_points.iloc[0]
    def_loss = budget_team.cum_points.iloc[d_idx] - players.cum_points.iloc[1]
    mid_loss = budget_team.cum_points.iloc[m_idx] - players.cum_points.iloc[2]
    fwd_loss = budget_team.cum_points.iloc[f_idx] - players.cum_points.iloc[3]

    # Add each loss to the net_loss list which is our unordered list of net losses
    net_loss.extend([gk_loss, def_loss, mid_loss, fwd_loss])

    # Create an ordered list of net losses high to low
    ordered_loss = net_loss.copy()
    ordered_loss.sort()

    #Â Now create an ordered list of indexes - i.e the index in net_loss that we can find values in the ordered_loss
    # Note the values have 1 added to equate to the element types 1=gk, 2=def, 3=mid, 4=fwd
    ordered_idx = [net_loss.index(ordered_loss[x])+1 for x in range(len(ordered_loss))]
    
    # Now we can iterate through the ordered_idx list and apply the value at that index as the index to find the next
    # most productive player in the current set. If they don't pass the cost test then we move onto the next
    
    for m, n in enumerate(ordered_idx):
        if n == 1 and players.iloc[0].cost < budget_team.iloc[g_idx].cost: #index 1 in base team == worst keeper
            budget_team = budget_team.drop(index=g_idx, axis=0)
            budget_team.loc[g_idx] = players.iloc[0]
            budget_team.sort_index(inplace=True)
            
        elif n == 2 and players.iloc[1].cost < budget_team.iloc[d_idx].cost: # index 6 in base team == worst defender
            budget_team= budget_team.drop(index=d_idx, axis=0)
            budget_team.loc[d_idx] = players.iloc[1]
            budget_team.sort_index(inplace=True)
        
        elif n == 3 and players.iloc[2].cost < budget_team.iloc[m_idx].cost: # index 11 in base team == worst midfield
            budget_team = budget_team.drop(index=m_idx, axis=0)
            budget_team.loc[m_idx] = players.iloc[2]
            budget_team.sort_index(inplace=True)
            
        elif n == 4 and players.iloc[3].cost < budget_team.iloc[f_idx].cost: # index 11 in base team == worst midfield
            budget_team = budget_team.drop(index=f_idx, axis=0)
            budget_team.loc[f_idx] = players.iloc[3]
            budget_team.sort_index(inplace=True)
            
            
    # If we have achieved a squad value equal or less than the Â£100M budget then we can break the loop    
    if budget_team.cost.sum()/10 < 100:
        break
    else:
        continue
    
```

#### Functions
1. Decide value and index of players up for transfer out of the squad, return the indexes
2. Create a scout dataframe that stores the players that we're looking to bring in, return the dataframe
3. Calculate the net loss of points each new transfer wold incur and return the ordered list of indexes
4. Create a master function that handles / calls to the above functions and then makes the final decision on transfer and then finally returns our completed best squad!

```python
#Â 1 decide on the index of players up for transfer out
def transfer_out(df):
    """
    df = base_team_sorted
    In this function we need to decide at what index, from the base squad, the player to be transfered is located at. 
    Typically, the indexes are set at the lowest scoring player location, however, could change!
    """
    

    # Set the baseline indexes to the location of poorest scoring player in the squad
    gk, dfn, mid, fwd = 1, 6, 11, 14
    
    # Check which players in our team are considered "low value" based on rules set out
    low_value = df[(df.cum_points<163) & (df.value<1.9)]
    low_value = low_value.sort_values(by="value", ascending=False)
    
    
    positions = df.groupby("element_type")
    
    for i in range(1, 5):
        pos = positions.get_group(i)
        new_transfer = pos[pos.web_name.isin(low_value.web_name)] 
        if i == 1 and new_transfer.shape[0]>0:
            gk = new_transfer.index.values[0]
        elif i == 2 and new_transfer.shape[0]>0:
            dfn = new_transfer.index.values[0]
        elif i == 3 and new_transfer.shape[0]>0:
            mid = new_transfer.index.values[0]
        elif i == 4 and new_transfer.shape[0]>0:
            fwd = new_transfer.index.values[0]
    
    
    return gk, dfn, mid, fwd
```

```python
def scout(df, i):
    """
    In this function we generate a dataset of players which include the next player in each position who is to be 
    considered for transfer into the squad
    df = df(full list of available players)
    i is the iterable value from the master function
    """
    
    # Split the full list of available players into position sets
    gk = df[df.element_type==1].sort_values(by="cum_points", ascending=False)
    dfn = df[df.element_type==2].sort_values(by="cum_points", ascending=False)
    mid = df[df.element_type==3].sort_values(by="cum_points", ascending=False)
    fwd = df[df.element_type==4].sort_values(by="cum_points", ascending=False)
    
    
    
    # Set up an empty dataframe to store the next players in each position
    targets = pd.DataFrame(columns = gk.columns)

    # Store each of the next players in an isolated frame
    next_gk = gk.iloc[i+2].to_frame().T
    next_def = dfn.iloc[i+5].to_frame().T
    next_mid = mid.iloc[i+5].to_frame().T
    next_fwd = fwd.iloc[i+4].to_frame().T

    # Add the players into the empty dataframe
    targets = targets.append(next_gk).append(next_def).append(next_mid).append(next_fwd)
    
    
    #Â Return the scouted players frame
    return targets
```

```python
def net_loss(df, targets, g_idx, d_idx, m_idx, f_idx):
    """
    In this function we calculate the potential loss of points of each player to be considered for transfer into 
    the squad and save them to a list. We then organise loss values in ascending order and create a list of 
    indexes that the ordered values are found in the initial loss list
    df = base_team_sorted
    """
    
    gk, dfn, mid, fwd = g_idx, d_idx, m_idx, f_idx
    
    # Set an empty list to store our player losses
    net_loss = []
    
    
    # calculate the potential loss each transfer would incur
    gk_loss = df.cum_points.iloc[gk] - targets.cum_points.iloc[0]
    def_loss = df.cum_points.iloc[dfn] - targets.cum_points.iloc[1]
    mid_loss = df.cum_points.iloc[mid] - targets.cum_points.iloc[2]
    fwd_loss = df.cum_points.iloc[fwd] - targets.cum_points.iloc[3]

    # Add each loss to the net_loss list which is our unordered list of net losses
    net_loss.extend([gk_loss, def_loss, mid_loss, fwd_loss])

    # Create an ordered list of net losses high to low
    ordered_loss = net_loss.copy()
    ordered_loss.sort()

    #Â Now create an ordered list of indexes - i.e the index in net_loss that we can find values in the ordered_loss
    # Note the values have 1 added to equate to the element types 1=gk, 2=def, 3=mid, 4=fwd
    ordered_idx = [net_loss.index(ordered_loss[x])+1 for x in range(len(ordered_loss))]
    
    return ordered_idx
```

```python
def generate_squad():
    """
    In this function we call to all our supporting functions, make the final decision on whether a player should
    be transfered, generate and return the best possible performing squad
    
    """
    best_team = base_team_sorted
    g_idx, d_idx, m_idx, f_idx = transfer_out(best_team)
    
    for i in range(gk.shape[0]-3): 
        
        players = scout(df, i) # Note df is the full list of available players
        
        ordered_idx = net_loss(best_team, players, g_idx, d_idx, m_idx, f_idx)
    
        for m, n in enumerate(ordered_idx):
            if n == 1 and players.iloc[0].cost < best_team.iloc[g_idx].cost: #index 1 in base team == worst keeper
                best_team = best_team.drop(index=g_idx, axis=0)
                best_team.loc[g_idx] = players.iloc[0]
                best_team.sort_index(inplace=True)

            elif n == 2 and players.iloc[1].cost < best_team.iloc[d_idx].cost: # index 6 in base team == worst defender
                best_team= best_team.drop(index=d_idx, axis=0)
                best_team.loc[d_idx] = players.iloc[1]
                best_team.sort_index(inplace=True)

            elif n == 3 and players.iloc[2].cost < best_team.iloc[m_idx].cost: # index 11 in base team == worst midfield
                best_team = budget_team.drop(index=m_idx, axis=0)
                best_team.loc[m_idx] = players.iloc[2]
                best_team.sort_index(inplace=True)

            elif n == 4 and players.iloc[3].cost < best_team.iloc[f_idx].cost: # index 11 in base team == worst midfield
                best_team = best_team.drop(index=f_idx, axis=0)
                best_team.loc[f_idx] = players.iloc[3]
                best_team.sort_index(inplace=True)

            
        # If we have achieved a squad value equal or less than the Â£100M budget then we can break the loop    
        if best_team.cost.sum()/10 < 100:
            break
        else:
            continue
    
    
    return best_team
```
